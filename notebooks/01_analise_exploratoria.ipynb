```python
import sys
import os
from pathlib import Path

# Adicionar raiz do projeto ao path
PROJECT_ROOT = Path(".").resolve().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

print(f"Project root: {PROJECT_ROOT}")

# Verificar arquivos de dados
PRESC_PATH  = PROJECT_ROOT / "data" / "raw" / "prescricoes.csv"
PARECERES_PATH = PROJECT_ROOT / "data" / "raw" / "pareceres.csv"

print(f"Prescrições:  {'✓' if PRESC_PATH.exists() else '✗ NÃO ENCONTRADO'} {PRESC_PATH}")
print(f"Pareceres:    {'✓' if PARECERES_PATH.exists() else '✗ NÃO ENCONTRADO'} {PARECERES_PATH}")
```

```python
from src.analysis.exploratory import load_csv, analyze_document_type, gerar_tabela1, gerar_graficos

OUTPUT_DIR = PROJECT_ROOT / "outputs" / "exploratory_analysis"
dfs = {}
stats = {}

for doc_type, path in [("prescricoes", PRESC_PATH), ("pareceres", PARECERES_PATH)]:
    if path.exists():
        print(f"Carregando {doc_type}...")
        df = load_csv(str(path))
        dfs[doc_type] = df
        print(f"  → {len(df):,} registros")
        stats[doc_type] = analyze_document_type(df, doc_type)
    else:
        print(f"⚠️  {doc_type}: arquivo não encontrado")
```

```python
# ── Tabela 1 ──────────────────────────────────────────────────────────────────
if stats:
    tabela1 = gerar_tabela1(stats, OUTPUT_DIR)
    display(tabela1)
```

```python
# ── Gráficos ──────────────────────────────────────────────────────────────────
if dfs:
    gerar_graficos(stats, dfs, OUTPUT_DIR)
    
    from IPython.display import Image
    display(Image(str(OUTPUT_DIR / "figura1_distribuicao_tokens.png")))
```

```python
# ── Inspeção manual de amostras ────────────────────────────────────────────────
# Visualizar exemplos de texto livre vs. template estruturado

if "prescricoes" in dfs:
    from src.analysis.exploratory import classify_text_type
    df_presc = dfs["prescricoes"].copy()
    df_presc["_type"] = df_presc["ds_evolucao"].apply(classify_text_type)
    
    print("=== EXEMPLO: Texto Livre ===")
    sample_livre = df_presc[df_presc["_type"] == "texto_livre"]["ds_evolucao"].dropna().iloc[0]
    print(sample_livre[:500])
    print()
    print("=== EXEMPLO: Template Estruturado ===")
    templates = df_presc[df_presc["_type"] == "template_estruturado"]["ds_evolucao"].dropna()
    if len(templates) > 0:
        print(templates.iloc[0][:500])
    else:
        print("(nenhum template encontrado na amostra)")
```
